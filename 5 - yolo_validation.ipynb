```python
from ultralytics import YOLO

model = YOLO("yolo12m.pt")
```

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = model.to(device)
```

per usare i pesi di COCO

```python
results = model.val(
    data="/content/dataset/data.yaml",
    imgsz= 1024
)

print(results)
```
