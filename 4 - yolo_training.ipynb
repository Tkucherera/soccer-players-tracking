```python
from ultralytics import YOLO

model = YOLO("yolo12m.pt")
```

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = model.to(device)
```

```python
model.train(
    data="/content/dataset/data.yaml",
    epochs=50,
    batch=16,
    imgsz=1024,
    device=device,                        # Usa GPU
    optimizer='AdamW',                    # Ottimizzatore stabile per fine-tuning
    lr0=3e-4,
    patience=15,                          # Early stopping
    workers=4,
    save=True,                            # Salva i checkpoint
    save_period=10,                       # Salva ogni 10 epoche
    resume=False,                         # Non riprende da un training precedente
    project="runs/train",                 # Directory dei risultati
    name="yolo12m",
    exist_ok=True,                        # Sovrascrive se la cartella esiste
    verbose=True,                         # Output dettagliato
    #freeze=10                            # Congela i primi 10 layer della rete
)
```
